```python
# begin by reading in the tourney outcomes data
to = pd.read_csv('tourney_outcomes.csv')
to
```

```python
# remove the datapoints that are not known in advance of the game
to_lean = to.drop(columns=['Score','OpponentScore'])
to_lean
```

```python
# now we can incorporate new data! lets add the regular season stats to each team, as well as the opponents stats
regseason = pd.read_csv('team_yearly_points.csv')
regseason
```

```python
# merge the dataframes on teamid and year
merged_data1 = pd.merge(to_lean, regseason, how='left', left_on=['TeamID', 'Season'], right_on=['TeamID', 'year'])
merged_data1
```

```python
merge2 = merged_data1.rename(columns={"Total Points":"RS Point Total","Average Points":"RS Avg Points"})
merge2
```

```python
merge2 = merge2.drop(columns=['year'])
merge2
```

```python
# now do the same for the opponent regular season stats
merge3 = pd.merge(merge2, regseason, how='left', left_on=['OpponentID', 'Season'], right_on=['TeamID', 'year'])
merge3
```

```python
merge4 = merge3.rename(columns={"Total Points":"OPP RS Point Total","Average Points":"OPP RS Avg Points"})
merge4
```

```python
merge5 = merge4.drop(columns=['TeamID_y','year'])
merge5
```

```python
merge6 = merge5.rename(columns={"TeamID_x":"TeamID"})
merge6
```

```python
merge6.tail(200)
```

```python
# how many rows do we lose if we drop nas?
merge7 = merge6.dropna()
merge7
```

```python
# now lets add a seed difference kpi
merge7['TeamSeedN'] = merge7['TeamSeed'].str.replace('[A-Za-z]', '')
merge7['OpponentSeedN'] = merge7['OpponentSeed'].str.replace('[A-Za-z]', '')

merge7['TeamSeedN'] = pd.to_numeric(merge7['TeamSeedN'])
merge7['OpponentSeedN'] = pd.to_numeric(merge7['OpponentSeedN'])

merge7['SeedDiff'] = merge7['TeamSeedN'] - merge7['OpponentSeedN']
```

```python
merge7
```

```python
# this is enough data for now, lets drop unneeded columns and start ml
```

```python
tourney_ml_output = merge7.drop(columns=['TeamSeed','OpponentSeed','TeamSeedN','OpponentSeedN'])
tourney_ml_output
```

```python
# write df to file so the DP notebooks don't have to be run again
tourney_ml_output.to_csv('tourney_ml_output.csv',index=False)
```

## Summary

* In this script we use the two datasets derived in DP1 (tourney_outcomes.csv) and DP2 (team_yearly_points.csv), and use them both to create a single file for use in ML.

* Before the join of tourney_outcomes.csv and team_yearly_points.csv I drop some features that would cause data leakage.

* After the join I drop rows that contain nulls. Some rows will have nulls as not all kaggle files have all data for all years. The Tournament results dataset only goes back to 1985, while the regular season detailed results file only goes back to 2003.

* After the rows with nulls are dropped we create a new kpi, seed difference.

* The final dataframe is exported to a csv, in a layout ready for ML.
